```python
import pandas as pd
import matplotlib.pyplot as plt

# Fix encoding issue
data = pd.read_csv("income_data.csv", encoding='latin-1')

# Bar chart
data['IncomeGroup'].value_counts().plot(kind='bar')

plt.title("Income Group Distribution")
plt.xlabel("Income Group")
plt.ylabel("Count")

plt.show()
```
